### Expérience 1 : Réseau Manhattan

```julia
# 1. Chargement des paramètres du réseau Manhattan
S, N, K, x, pos, lambda, sigma, t_velo, t_camion = manhattan()

# 2. Exécution de la simulation sur 50 000 unités de temps
sim_time = 50000.0
resultats_manhattan = run_simulation(S, N, K, x, lambda, sigma, t_velo, t_camion, sim_time)

# 3. Affichage de la performance : Usagers satisfaits en fonction du temps
plot(
    resultats_manhattan[:liste_t], 
    resultats_manhattan[:liste_k],
    title = "Performance du camion - Manhattan",
    xlabel = "Temps de simulation",
    ylabel = "Usagers satisfaits (k)",
    label = "k(t)",
    linewidth = 2,
    color = :blue
)
```

### Expérience 2 : Réseau Asymétrique

```julia
# 1. Chargement des paramètres du réseau Trio Asymétrique
S_a, N_a, K_a, x_a, pos_a, lambda_a, sigma_a, t_velo_a, t_camion_a = trio_asymetrique()

# 2. Exécution
resultats_asym = run_simulation(S_a, N_a, K_a, x_a, lambda_a, sigma_a, t_velo_a, t_camion_a, sim_time)

# 3. Affichage de la comparaison
plot(
    resultats_asym[:liste_t], 
    resultats_asym[:liste_k],
    title = "Performance du camion - Topologie Asymétrique",
    xlabel = "Temps de simulation",
    ylabel = "Usagers satisfaits (k)",
    label = "k(t) Asymétrique",
    linewidth = 2,
    color = :red
)
```
